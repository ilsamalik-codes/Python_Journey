```python
def calculate_bill(cart, prices):
    total_bill = 0
    item_costs = {}
    most_expensive_item = None
    max_price = -1

    for item, qty in cart.items():
        if item not in prices:
            print("Warning:", item, "not found in price list, ignored.")
            continue

        unit_price = prices[item]
        cost = qty * unit_price

        item_costs[item] = cost
        total_bill += cost

        if unit_price > max_price:
            max_price = unit_price
            most_expensive_item = item

    return total_bill, most_expensive_item, item_costs


cart = {"apple": 4, "banana": 6, "milk": 2, "bread": 1}
prices = {"apple": 30, "banana": 10, "milk": 55, "bread": 40}
cart["eggs"] = 12  # Adding an item not in prices to test warning
total_bill, most_expensive_item, item_costs = calculate_bill(cart, prices)

print("Total bill:", total_bill)
print("Most expensive item:", most_expensive_item)
print("Item-wise costs:")
for item, cost in item_costs.items():
    print(item + ":", cost)
```
